```python
import pandas as pd
import matplotlib.pyplot as plt
import pickle
import numpy as np
import math
from matplotlib import collections  as mc
```

```python
%matplotlib inline

import sys, os
```

```python
# Load the dataframe
df0 = pd.read_csv('test_data/logs/1.csv', sep=',', names= ['energy', 'size'],skiprows=1)
df1 = pd.DataFrame({'timestep': range(1,1+df0.index.size)}) 

# Set temp schedule to match generated results
startingTemp = 50000
finalTemp = 0.1
Tfactor = -math.log(startingTemp / finalTemp)
steps = df0.shape[0]
schedule = pd.Series([startingTemp * math.exp(Tfactor * step / steps) for step in range(0,steps)])

print((schedule).shape)
df2 = pd.DataFrame({'temperature': schedule }) 
df3 = df0.join(df1)
df3 = df3.join(df2)

# Initialize BIG figure
fig = plt.figure(1, [20, 4])   

plt.rcParams.update({'font.size': 8})

# Three subplots, the axes array is 1-d
f, axarr = plt.subplots(4, sharex=False)

# I load the seaborn library only to benefit the nice looking feature
import seaborn as sns
axarr[0].text(-50000, 1.0*max(df3['temperature']), 'A', size=20, weight='bold')
axarr[0].plot('timestep','temperature',data=df3.iloc[::1000, :],color='grey',alpha=.4,lw=3.5)
axarr[0].set_ylabel('Simulation\ntemperature\n(mmol/gDW/h)')

axarr[1].text(-50000, 1.0*max(df3['energy']), 'B', size=20, weight='bold')
axarr[1].semilogy('timestep','energy',data=df3.iloc[::100, :],color='skyblue', linewidth=2.0)
axarr[1].set_yticks([1,10,100])
axarr[1].set_ylabel('Energy, e.g.\nflux into core\n(mmol/gDW/h)')
#axarr[1].plot('timestep','energy',data=df3.iloc[::150, :],color='black',alpha=.1,lw=.1)

axarr[2].text(-50000, 1.0*max(df3['size']), 'C', size=20, weight='bold')
axarr[2].plot('timestep','size',data=df3.iloc[::1000, :],color='red',alpha=.8,lw=1.8)
#axarr[2].plot('timestep','size',data=df3.iloc[::150, :],color='black',alpha=.1,lw=.1)
axarr[2].set_xlabel('Simulation Timestep (subsampled)')
axarr[2].set_ylabel('Core size\n (reactions)')

#plot zoom lines
axarr[2].plot(*[(35000, 35000), (127, 143), 'black'])
axarr[2].plot(*[(39000, 39000), (127, 143), 'black'])

axarr[3].text(34000, 1.0*max(df3['size']), 'D', size=20, weight='bold')
axarr[3].plot('timestep','size',data=df3.iloc[35000:39000],color='red',alpha=.8,lw=1.8)
#axarr[3].plot('timestep','size',data=df3.iloc[::150, :],color='black',alpha=.1,lw=.1)
axarr[3].set_xlabel('Simulation Timestep (raw)')
axarr[3].set_ylabel('Core size\n (reactions)')
#axarr[3].set_xlim([50000,75000])

plt.tight_layout()
plt.savefig("fig4-v7.pdf", bbox_inches='tight')
```
