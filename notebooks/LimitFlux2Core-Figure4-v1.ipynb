```python
# Load the dataframe
df1 = pd.read_csv('test_data/196.csv', sep=',', names= ['energy', 'size'],skiprows=1)
df2 = pd.DataFrame({'timestep': range(1,1+df1.index.size)}) 
df3 = df1.join(df2)

# Initialize BIG figure
fig = plt.figure(1, [8, 4])    

# Two subplots, the axes array is 1-d
f, axarr = plt.subplots(2, sharex=True)

# I load the seaborn library only to benefit the nice looking feature
import seaborn as sns
axarr[0].plot('timestep','energy',data=df3,color='skyblue')
axarr[0].set_ylabel('Flux into core \n(mmol/gDW/h)')
axarr[0].plot('timestep','energy',data=df3,color='black',alpha=.1,lw=.1)

axarr[1].plot('timestep','size',data=df3,color='red',alpha=.8,lw=1)
axarr[1].plot('timestep','size',data=df3,color='black',alpha=.1,lw=.1)

axarr[1].set_xlabel('Simulation Timestep')
axarr[1].set_ylabel('Core size\n (reactions)')
axarr[1].set_ylim([-1,175])

plt.tight_layout()
plt.savefig("fig4-v1.pdf", bbox_inches='tight')
```
